```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import ElementClickInterceptedException
from time import sleep
```

```python
import pandas as pd
```

```python
def kliknij_maile():    
    xpath = "(//button[@type='button'][contains(text(),'Pokaż email')])"
    elements = []
    elements = WebDriverWait(d, 20).until(
        EC.presence_of_all_elements_located((By.XPATH, xpath)))
    for element in elements:
        element.click()
```

```python
d = webdriver.Chrome(r"C:\Users\User\Pythom\chromedriver-win64\chromedriver.exe")
site = 'https:/www.prawo-jazdy-360.pl/ranking-szkol-jazdy/warszawa'
d.get(site)
```

```python
dictionary = {}
i=2
for nr_strony in range (2,22):
    print (f"Przetwarzam stronę {nr_strony}")
    try:
        kliknij_maile()
    except ElementClickInterceptedException as e:
        print(f"Wystąpił błąd ElementClickInterceptedException: {e}")
    mails=[]
    ms=d.find_elements(By.PARTIAL_LINK_TEXT, "@")
    for m in ms:
        mails.append(m.text)
        
    titles=[]
    ts=d.find_elements(By.CLASS_NAME, 'info')
    for t in ts:
        titles.append(t.text)
    for t, m in zip(titles[:10], mails[:10]):
        dictionary[t] = m
    
    if nr_strony < 21:        
        page_xpath = f"//a[normalize-space()='{nr_strony}']"
        next_page_button = d.find_element(By.XPATH, page_xpath)
        next_page_button.click()
        # w8 for page 
        sleep(6)       
d.close()
```

```python
df = pd.DataFrame.from_dict(dictionary, orient = 'index', columns=['Nazwa Ośrodka', 'Mail'])
```
